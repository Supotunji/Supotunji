```python
# calculate the Chi Square using only non-numeric columns to determine appropriate columns to drop
# List of non-numeric columns to analyze
non_numeric_columns = [
    'NAME_CONTRACT_TYPE', 'NAME_TYPE_SUITE', 'NAME_INCOME_TYPE',
    'NAME_EDUCATION_TYPE', 'NAME_FAMILY_STATUS', 'NAME_HOUSING_TYPE',
    'OCCUPATION_TYPE', 'WEEKDAY_APPR_PROCESS_START', 'ORGANIZATION_TYPE'
]

# Initialize a dictionary to store results
chi2_results = {}

# Iterate through non-numeric columns
for column in non_numeric_columns:
    # Create a contingency table for the current column and 'TARGET'
    chi_table = pd.crosstab(applicationdata_cl[column], applicationdata_cl['TARGET'])

    # Perform chi-square test
    chi2, p_value, dof, expected = chi2_contingency(chi_table)

    # Store results in the dictionary
    chi2_results[column] = {'Chi-square': chi2, 'P-value': p_value}

# Print the chi-square test results
for column, result in chi2_results.items():
    print(f"Column: {column}")
    print(f"Chi-square test statistic: {result['Chi-square']}")
    print(f"P-value: {result['P-value']}")
    print("")
```
